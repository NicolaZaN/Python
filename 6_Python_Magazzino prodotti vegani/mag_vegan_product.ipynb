```python
"""

Software store of vegan product

"""

import json

import sys

import os

 

"""Function to add products to the warehouse"""


def add_product(shop, q):

    while True:
        
        name_product = input("Name of product: ")
        
        if not name_product.isalpha():
            
            print("Enter a correct name of product")
            continue
            
        break

    while True:
            
        try:
            q_tà = int(input("Quantity: "))
            
            if q_tà < 0:
                
                print("Enter a positive quantity")
                
                continue

        except ValueError:

                print("Enter a correct quantity")

                continue
                
        break  

    q += q_tà


    if name_product.lower() not in shop:

        while True:
        
            aq_price = input("Enter a purchase price: ")
            
            if  aq_price.isalpha():
            
                print("Enter a correct purchase price")
                continue
        
            if int(aq_price) < 0:
            
                print("Enter a price > 0")
                continue
                
            break
        
        aq_price = float(aq_price)
                
        while True:

            sell_price = input("Enter a selling price: ")
            
            if  sell_price.isalpha():
            
                print("Enter a correct selling price")
                continue
        
            if int(sell_price) < 0:
            
                print("Enter a price > 0")
                continue
                
            break
        
        sell_price = float(sell_price)
            
        list_val_product = []

        list_val_product.append(q_tà)

        list_val_product.append(aq_price)

        list_val_product.append(sell_price)

    
        shop[name_product] = list_val_product

    else:

        shop[name_product][0] += q_tà
    
    print(f"\nAdd {q_tà} X {name_product}")

    return q

 

"""Function that allows you to show the list of products in the warehouse"""

 

def show_products(shop):

    product_keys = list(shop.keys())

    product_values = list(shop.values())


    print("PRODUCT QUANTITY' PRICE")

    for i in range(len(shop.keys())):

        print(product_keys[i] + " " + str(product_values[i][0]) + " " + "€" + str(product_values[i][2]))

    return

 

"""Function that allows you to sell the products in the warehouse"""


def sell_product(shop):


    dict_to_sell = {}

    list_to_sell = []
    
    tot_profit = 0
    
    tot_net = 0

    shop = clear_shop(shop)
    
    if shop == {}:
        
        print("Store si empty")
    
    else:
        
        while True:

            p_to_sell = input("Name of product: ")

            if p_to_sell not in shop.keys():

                print("The product is not in stock, enter a product in stock")

                continue

            while True:

                q_to_sell = input("Quantity: ")

                if q_to_sell.isalpha():

                    print("Enter a number")

                    continue
                if int(q_to_sell) < 0:

                    print("Enter positive number")

                    continue

                q_to_sell = int(q_to_sell)

                if q_to_sell > shop[p_to_sell][0]:

                    print("The quantity entered is too high")

                    continue

                break

            product_profit = q_to_sell*shop[p_to_sell][2]

            product_net = product_profit - (q_to_sell*shop[p_to_sell][1])

            shop[p_to_sell][0] -= q_to_sell

            list_to_sell.append(q_to_sell) #Enter quantity of product into selling list

            list_to_sell.append(shop[p_to_sell][2]) #Enter selling price of product to sell

            list_to_sell.append(product_profit) #Enter product's proftit

            list_to_sell.append(product_net) #Enter product's net

            if p_to_sell not in dict_to_sell:

                dict_to_sell[p_to_sell] = list_to_sell

            else:

                dict_to_sell[p_to_sell][0] += q_to_sell
                dict_to_sell[p_to_sell][2] += product_profit
                dict_to_sell[p_to_sell][3] += product_net
                 
            p = input("Do you want to add another product(yes/no): ")

            if p == "yes":
                
                c = []
                
                for i in shop.values():
                    
                    c.append(i[0])
                    
                if sum(c) == 0:

                    print("Store is empty")
                    
                    break
                        
                else:  
                        
                    continue
                        
            elif p != "no":
                
                print("Error command not recognized ")
            
            elif p == "no":
                
                break
                    
    print("\nREGISTERED SALE")
    tot_profit = 0
    tot_net = 0

    for product in dict_to_sell.items():

        print(f"- {product[1][0]} X {product[0]}: €{product[1][1]}")

        tot_profit += product[1][2]
        tot_net += product[1][3]

        print(f"TOT: €{tot_profit}")

        break
            
    if os.path.isfile("profit_and_net_total.txt"):
        
        with open("profit_and_net_total.txt", "r", encoding ="utf-8") as profit_net_file:
        
            profit_net_file.seek(0)
        
            k = profit_net_file.readline()
            j = profit_net_file.readline()
        
            if k != "":
            
                tot_profit += float(k)
                tot_net += float(j)
        
    with open("profit_and_net_total.txt", "w", encoding ="utf-8") as profit_net_file:
        
        profit_net_file.write(str(tot_profit))
        profit_net_file.write("\n")
        profit_net_file.write(str(tot_net))
            
    return tot_profit, tot_net, dict_to_sell

       

def total_profits(tot_profit, tot_net, dict_to_sell):

    for i in dict_to_sell:

        tot_profit += dict_to_sell[i][2]

        tot_net += dict_to_sell[i][3]


    return tot_profit, tot_net, dict_to_sell

           
def clear_shop(shop):


    lista_shop = list(shop.items())

    for i in shop.copy():

        if shop[i][0] == 0:

            shop.pop(i) 

    return shop

   



"""Program start"""

if os.path.isfile("store.json"):

    with open("store.json", "r", encoding ="utf-8") as mag_file:

        shop = json.load(mag_file)

else:

    shop = {}
    
command = input("Enter a command: ")

print("\n")

while command.lower() != "close":
    if command.lower() == "help" or command.lower() not in ["add", "list", "sale", "profits", "close"] :

        print("""The available commands are the following:

                                        add: add a product to the warehouse

                                        list: lists the products in stock

                                        sale: records a sale made

                                        profits: Shows total profits

                                        help: shows possible commands

                                        close: exit the program""")   

    if command.lower() == "exit":

        sys.exit("Program close")
        
    if command.lower() == "add":

        q = 0
        q = add_product(shop,q) 
                        
    if command.lower() == "list":
             
        clear_shop(shop)

        show_products(shop)


    if command.lower() == "sale":
        
        with open("store.json", "r") as mag_file:
            line = json.load(mag_file)
        
        if line == {}:
            
            print("Impossible to sell: there are no products in stock")
        
        else:
            
            tot_profit, tot_net, dict_to_sell = sell_product(shop)
        

    if command.lower() == "profits":
        
        dict_to_sell = {}
        
        tot_profit = 0
        tot_net = 0
        
        tot_profit, tot_net, dict_to_sell = total_profits(tot_profit, tot_net, dict_to_sell)
        
        if not os.path.isfile("profit_and_net_total.txt"):
                    
                print("Gross profit: = 0 net: = 0")

        else:
            
            with open("profit_and_net_total.txt", "r", encoding ="utf-8") as profit_net_file:
                
                profit_net_file.seek(0)
                k = profit_net_file.readline()
                j = profit_net_file.readline()
                print("Profit:\n")
                print(f"gross: = {k}")
                print(f"net: = {j}")
                profit_net_file.seek(0)
                
                    
            with open("profit_and_net_total.txt", "w", encoding ="utf-8") as profit_net_file:
                
                profit_net_file.seek(0)
                profit_net_file.write(k)
                profit_net_file.write(j)     
                profit_net_file.seek(0)
                    
            
    command = input("\nEnter a command: ")


    with open("store.json", "w", encoding ="utf-8") as mag_file:

        json.dump(shop, mag_file)

    with open("store.json", "r", encoding ="utf-8") as mag_file:

        shop = json.load(mag_file)
        
clear_shop(shop)
```
